```python
from matplotlib.colors import ListedColormap
from sklearn import cross_validation, metrics, neighbors 
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.datasets import load_digits, load_breast_cancer
from pandas import DataFrame
import matplotlib.pyplot as pl

import numpy as np
%matplotlib inline
```

```python
cancer = load_breast_cancer()
cancer_frame = DataFrame(cancer.data)
cancer_frame.columns = cancer.feature_names
cancer_frame['target'] = cancer.target
```

```python
cancer_frame.head()
```

```python
for dataset, d_name in zip([digits, cancer], ["digits", "cancer"]):
    for estimator, e_name in zip([GaussianNB(), MultinomialNB(), BernoulliNB()], 
                             ["gauss", "multinomial", "bernoulli"]):
        print("dataset: {}, distrubution: {}, accuracy: {}".format(d_name, 
              e_name, (cross_val_score(estimator, dataset['data'], dataset['target'])).mean()))
```

# 1) Для $breast\ cancer$ максимальный $accuracy$ равен $0,937$
# 2) Для $digits$ максимальный $accuracy$ равен $0,871$
# 3) Верно $c, d$
# Пояснение к a,b: в случае $cancer$ (вещественные признаки) наилучший $accuracy$ у классификатора с нормальным распределением.
# Чем ближе распределение, используемое в классификаторе, к настояющему распределению выборки признаков, тем лучше (верее лучше с большей вероятностью, ведь по-другому и не получится) будет проведена классификация. 
# В случае $digits$ признаком был массив из $64$ чисел от $0$ до $16$. Неудивительно, что мультиномиальное распределение лучше других описывает распределение этого признака.
# В случае $cancer$ признаки вещественные, поэтому, опять же, не стоит удивляться, что нормальное распределение показало себя лучше других.
