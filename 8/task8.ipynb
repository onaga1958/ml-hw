Реализуйте PCA

```python
 '''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def myPCA(F, varRetained=0.95, show=False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n, l) = F.shape
    Sigma = 1.0 / l * np.dot(F, np.transpose(F))
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    U, d, V = np.linalg.svd(F.T) # посчитать SVD матрицы Sigma
    # U, d, V = np.linalg.svd(Sigma)
    # compute the value m: number of minumum features that retains the given variance varRetaine
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[:i + 1]) / dTot * 100.0 for i in range(n)])
    m = np.argmax(var_i > (varRetained * 100))
    print('{:.2f} % variance retained in {} dimensions'.format(var_i[m], m))
    
    # plot the variance plot
    if show:
        pl.plot(var_i)
        pl.xlabel('Number of Features')
        pl.ylabel(' Percentage Variance retained')
        pl.title('PCA $\% \sigma^2 $ vs # features')
        pl.show()

    # compute the reduced dimensional features by projection
    U_reduced = V.T[:, :m]
    G = np.dot(F.T, U_reduced)

    return G, U_reduced
```

```python
# Примените алгоритм к данным MNIST
from sklearn.model_selection import train_test_split

from sklearn.datasets import load_digits
X, y = load_digits(return_X_y=True)
X_train, X_test, Y_train, Y_test = train_test_split(X, y, train_size=0.7)
```

```python
#################################################################
# PCA of training set
print('Performing PCA - Principal COmponent Analysis')

G, U_reduced = myPCA(X.T, varRetained = 0.95, show = True)
```

Теперь займемся кластеризацией

```python
def make_dbscan(eps, min_samples, need_plot, need_print):
    # beautiful plot from here
    # http://scikit-learn.org/stable/auto_examples/cluster/plot_dbscan.html#sphx-glr-auto-examples-cluster-plot-dbscan-py
    db = DBSCAN(eps=eps, min_samples=min_samples).fit(X_hat)
    core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
    core_samples_mask[db.core_sample_indices_] = True
    labels = db.labels_

    # Number of clusters in labels, ignoring noise if present.
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    result = (metrics.adjusted_rand_score(target, labels),
              metrics.adjusted_mutual_info_score(target, labels))
    
    if need_print:
        print('Estimated number of clusters: %d' % n_clusters_)
        print("Adjusted Rand Index: %0.3f"
              % result[0])
        print("Adjusted Mutual Information: %0.3f"
              % result[1])
    
    if need_plot:
        unique_labels = set(labels)
        colors = pl.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
        pl.figure(figsize=(12, 12))
        
        for k, col in zip(unique_labels, colors):
            if k == -1:
                # Black used for noise.
                col = 'k'

            class_member_mask = (labels == k)

            xy = X_hat[class_member_mask & core_samples_mask]
            pl.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
                    markeredgecolor='k', markersize=14)

            xy = X_hat[class_member_mask & ~core_samples_mask]
            pl.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
                    markeredgecolor='k', markersize=6)
        pl.show()
            
        for k in unique_labels:
            class_member_mask = (labels == k)

            if k == -1:
                print('emissions')
                size = min(len(X[class_member_mask]), 10)
                images = random.sample(list(X[class_member_mask]), size)
            else:
                print('core points of {} cluster'.format(k))
                size = min(len(X[class_member_mask & core_samples_mask]), 10)
                images = random.sample(list(X[class_member_mask & core_samples_mask]), 10)

            pl.figure(figsize=(3 + size, 1))
            for j, image in enumerate(images):
                pl.subplot(1, size, j + 1)
                pl.imshow(image.reshape((8, 8)), cmap='gray')
                pl.axis('off')

            pl.show()
            
    return result
```

```python
def get_best_params(epsilons, min_samples, need_print):
    results = []

    all_variants = list(product(epsilons, min_samples))
    for eps, min_s in all_variants:
        if need_print:
            print(eps, min_s)
        results.append(make_dbscan(eps, min_s, False, need_print))

    ind = np.argmax([x[0] + x[1] for x in results])
    return all_variants[ind]
```

```python
def reduce_colors(image, method, **kwargs):
    reshaped_image = np.copy(image).reshape(-1, 3)
    
    if method == 'meanshift':
        bandwidth = estimate_bandwidth(reshaped_image, n_samples=1000, **kwargs)
        clusterizator = MeanShift(bandwidth=bandwidth, bin_seeding=True)
    elif method == 'kmeans':
        clusterizator = KMeans(**kwargs)
    elif method == 'agglomerative':
        connectivity = kneighbors_graph(reshaped_image, 30)
        clusterizator = AgglomerativeClustering(connectivity=connectivity, **kwargs)
    else:
        raise ValueError('Unknown method: ' + method)
    
    result = clusterizator.fit_predict(reshaped_image)
    clusters_num = len(np.unique(result))
    
    for cluster_ind in range(clusters_num):
        cluster_mask = result == cluster_ind
        curr_cluster = reshaped_image[cluster_mask]
        reshaped_image[cluster_mask] = np.mean(curr_cluster, axis=0)

    new_image = reshaped_image.reshape(image.shape)
    
    print("method: {}, K: {}".format(method, clusters_num))
    print("SSIM: {}".format(compare_ssim(image, new_image, multichannel=True)))
    pl.figure(figsize=(8, 6))
    pl.imshow(new_image)
    pl.axis('off')
    pl.show()
```

```python
clusters_nums = [2, 3, 10, 20]
kwargs = [['n_clusters', [2, 3, 10, 20]],
          ['n_clusters', [2, 3, 10, 20]],
          ['quantile', [0.2, 0.15, 0.1, 0.05, 0.025]]]
methods = ['agglomerative', 'kmeans', 'meanshift']

for method, kwarg in zip(methods, kwargs):
    for value in kwarg[1]:
        reduce_colors(image, method, **{kwarg[0]:value})
```
